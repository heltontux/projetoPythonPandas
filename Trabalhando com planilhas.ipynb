```python
#Importando a biblioteca
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datasets/Aracaju.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datasets/Fortaleza.xlsx')
df3 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datasets/Natal.xlsx')
df4 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datasets/Recife.xlsx')
df5 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datasets/Salvador.xlsx')
```

```python
df5.head()
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
# Exibindo as 5 primeiras linhas
df.head()
```

```python
# EXibindo a5 ultimas linhas 
df.tail()
```

```python
# Escolhe uma linha aleatória
Se quiser exibir mais de uma linha basta colocar dentro do colchetes
df.sample()
```

```python
df.sample(15)
```

```python
# Verificando os tipos de dados
# Observe que este comando é sem parenteses
df.dtypes
```

```python
# A coluna LojaID é do tipo inteiro, mas como esta coluna não servirá para cálculo,
# podemos tratá-la como String (O pandas entende como object)
#
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df['LojaID'].astype('object')
```

```python
df.dtypes
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela média
# O parâmetro inplace=True altera o arquivo na origem
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
# Apagando as linhas com valores nulos por zero
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
#Criando a coluna de receita
# mul => multiplicar
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
# Criando a coluna Receita por Venda
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
df.head()
```

```python
# nlargest => Retorna o maior valor da columna passada por parâmetro.
# obrigatório colocar o numero de linhas que se quer exibir.
df.nlargest(3, "Receita")
```

```python
# nsmallest => Retorna o menor valor da columna passada por parâmetro.
# obrigatório colocar o numero de linhas que se quer exibir.
df.nsmallest(3, "Receita")
```

```python
# Soma das receitas por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
# Ordenando por receita.
# Se for omitida a ordenação, por padrão é ordenado do menor para o maior (ascending = True)
df.sort_values("Receita", ascending=False).head(10)
```
